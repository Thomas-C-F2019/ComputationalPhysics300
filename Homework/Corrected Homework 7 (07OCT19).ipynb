```python
#Thomas Conibear - Online Data Compression Assignment
#This is the correct homework. I did the initial assignment incorrectly


import urllib.request
import matplotlib.pyplot as plt
from scipy.fftpack import fft
import numpy
from numpy import fft


#The urllib.request method from notes did not work for me so I used this method
#Extracting data from online website

OnlineData = numpy.loadtxt("http://www-personal.umich.edu/~mejn/cp/data/circular.txt")


#Perform fast fourier analysis
Fourier = numpy.fft.fft2(OnlineData)
Fourier2 = numpy.fft.fftshift(Fourier) #Must move data to make it fit our graph
FourierGraph = 20*numpy.log(numpy.abs(Fourier2)) #Makes it more suitable for graph

plt.figure(figsize=[12, 9])


#Graph uncompressed data to compare
plt.imshow(OnlineData, cmap = "inferno")
plt.title("Original Uncompressed Data")
plt.colorbar()
plt.show()

plt.figure(figsize=[12, 9])

plt.imshow(FourierGraph, cmap = "inferno") #Take absolute value to eliminate imaginary outputs
plt.title("Compressed Data After Performing Fourier Transformation")
plt.colorbar()
plt.show()
```

```python
#Thomas Conibear - Online Data Compression Assignment


import urllib.request
import matplotlib.pyplot as plt
from scipy.fftpack import fft
import numpy


#The urllib.request method from notes did not work for me so I used this method
#Extracting data from online website

OnlineData = numpy.loadtxt("http://www-personal.umich.edu/~mejn/cp/data/circular.txt")


#Perform fast fourier analysis
Fourier = fft(OnlineData)

plt.figure(figsize=[12, 9])

#Graph uncompressed data to compare
plt.plot(OnlineData)
plt.title("Original Uncompressed Data")
plt.show()

plt.figure(figsize=[12, 9])

plt.plot(abs(Fourier)) #Take absolute value to eliminate imaginary outputs
plt.title("Compressed Data After Performing Fourier Transformation")
plt.show()
```

As you can see the fast Fourier transformation greatly reduced the sheer scope of the data into several manageable and relevant sets of data that follow a curved pattern. There are about four sets or lines of data that the Fourier transformation has worked out as important to us from the incredibly noisy data that we originally had in the first graph. This demonstrates how a Fourier Transformation can convert thousands of inconsistent data into several useful lines or sets of data to analyze.

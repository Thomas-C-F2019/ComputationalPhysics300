```python
import numpy

N = 1000000 #Number of rolls

Double_Six = 0 #Counting how many times double six is rolled, start at 0

for i in range(N):
    x = numpy.random.randint(1, 7, 1) 
    y = numpy.random.randint(1, 7, 1)
    if (x == 6) and (y == 6):
        Double_Six += 1
        
z = Double_Six / N  #Calculating probability

print("The number of times a double six is rolled: " , Double_Six)
print("The probability of rolling a double six is: " , z)
```

```python
import matplotlib.pyplot as plt

#I use the random number generator code from the notes but the magic numbers there are general magic numbers
#They are not good for constraining the random integers from -5 to 5
#So we have to find new magic numbers that achieve this. It seems that the only way to do this is guess and check

N = 1000
a = 53 #16807
c = 73*1/11 #1012
m = 10 #
x = 1

results = []

for i in range(N):      
    x = (a*x+c)%m
    results.append(x-5) #Need to do minus five to achieve -5 to 5 as we can't obtain negative values with our equation
    
plt.plot(results,"o")
plt.show()

plt.hist(results)
plt.show()
```

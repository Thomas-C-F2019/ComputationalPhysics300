```python
#Thomas Conibear - Homework #4 on Integrals


#First integral f(x)= (x**2 - x + 1) from 0 to 3

import numpy
import matplotlib.pyplot as plt

def Rectangle1(start, end, parts, plot=True):
    
    #define the function we want to integrate
    f = lambda x: (x**2 - x + 1)
    
    #define x and y points. Parts are the bins/rectangles we want. Results x gives you width but you don't know 
    #height, thats what results y is for
    deltax = (end - start) / parts
    resultsx = numpy.linspace(start, end, parts)
    resultsy = f(resultsx)
    
    #Calculating area
    area = numpy.empty([parts], float)
    for i in range(parts):
        area[i] = resultsy[i] * deltax
    
    if plot:
        x = numpy.linspace(start,end,100)
        y = (x**2 - x + 1)
        plt.title("Rectangle Method for First Integral")
        plt.plot(x, y, 'r')
        plt.xlim([start,end])
        plt.bar(resultsx+deltax/2, resultsy, deltax, edgecolor ='black')
        plt.show()
        print("The Area of the Rectangle Method is: ", sum(area))
        
    return sum(area)

Rectangle1(0,3,200)


def Trapezoid1(start, end, parts, plot=True):
    
    f = lambda x: (x**2 - x + 1)  
     
    deltax = (end - start) / parts
    resultsx = numpy.linspace(start, end, parts+1)
    resultsy = f(resultsx)

    # To calculate the area
    area = numpy.empty([parts], float)
    for i in range(parts):
            area[i] = (resultsy[i]+resultsy[i+1]) * deltax/2

    if plot==True:
        x = numpy.linspace(start,end,100)
        y = f(x)        
        plt.plot(x, y, 'r')
        plt.xlim([start,end])
        
        y2 = numpy.array([0,0])
        for i in range(parts):
            x0 = resultsx[i:i+2]
            y1 = resultsy[i:i+2]
            plt.fill_between(x0, y1, y2, where=y1>=y2, facecolor='blue')
            linex, liney = [resultsx[i+1], resultsx[i+1]], [0, resultsy[i+1]]
            plt.plot(linex, liney, color='black')        
            plt.title("Trapezoid Method for First Integral")
        
        plt.show()
        print("The Area of the Trapezoid Method is: ", sum(area))
    
    return sum(area)

Trapezoid1(0,3,20)


def Simpson1(start, end, parts):
    f = lambda x: (x**2 - x + 1)
    
    deltax = (end - start) / parts
    resultx = numpy.linspace(start, end , parts + 1)
    resulty = f(resultx)
    
    area = numpy.empty([parts], float)
    
    for i in range(parts):
        if i == 0 or i == parts:
            area[i] = 1/3 * resulty[i] * deltax
        elif i % 2 == 0:
            area[i] = 2/3 * resulty[i] * deltax
        elif i % 2 != 0:
            area[i] = 4/3 * resulty[i] * deltax

        print("The Area of the Simpson Method is: ", sum(area))
        return sum(area)
Simpson1(0,3, 20)
```

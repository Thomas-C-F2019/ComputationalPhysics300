```python
#Thomas Conibear - Curve Fitting Homework
  

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=[12, 10])

#define the function
f = lambda x: (x+x+x)**3

#define the paramaters for the plot
a,b = -0.75,0.75
npoints = 100
order = 4

x = np.linspace(a,b,npoints)
y = f(x) + np.random.rand(npoints) - 0.7

fit = np.polyfit(x, y, order, full=True)
y_p = np.polyval(fit[0], x)

plt.plot(x,y, 'o', label='Noisy Data')
plt.plot(x,f(x), label='Ideal Function')
plt.plot(x,y_p, label='Polynomial Fit Order '+str(order))

plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.xlim([a,b])
plt.legend(fontsize=12)
plt.title("Data Fitting Given Noisy Data")
plt.show()
print('Fitted Coefficients: ' + str(fit[0]) + '\n Residual: ' + str(fit[1]))
```

The residual value of our fit and graph indicates the accuracy of our f(x) values to the fitted function. So a lower residual value indicates more consistency between our data and our fitted function, while a higher residual value indicates the fit and data are not consistent. By analyzing the residual of the fit and the look or curve of the fit, I was able to conclude that the ideal order of the fit to my chosen function is order 3 or 4. 

Because fitting doesn't work well for a large range of noisy data points, I zoomed in locally to an interesting part of this function and analyzed it with a fit. 

Removing - 0.5 in the randomness led to all random points being placed above the curve so I played around with it and found that - 0.7 seems to give the most even amount of random points above and below the curve in this case.

Because of this I conclude that the orders of 3 and 4 are the most ideal fit to the function I specified above. These two orders give the best fits because their order is identical or very close to the order of the original function. The fit is capable of accurately mapping a function to the points when it's order is on or near the magnitude of the original function being used to create the noisy data that we are fitting.

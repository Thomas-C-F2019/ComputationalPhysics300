```python
#Thomas Conibear - One-Rule Machine Learning Homework

#From notes we know that it is easier for a computer to find a rule if all values are one or zero
#From notes we have the simple code to achieve this

from sklearn.datasets import load_iris
import numpy as np

dataset = load_iris() #Data from sklearn library
X = dataset.data
Y = dataset.target

X_d = np.array( X>X.mean(), dtype='int') #Converts to ones and zeroes using mean values
```

```python
#From notes to get a table of ones and zeroes
from collections import OrderedDict
import pandas as pd

Iris_data_d = OrderedDict({'Sepal Length ': X_d[:,0],
             'Sepal Width': X_d[:,1],
             'Petal Length': X_d[:,2],
             'Petal Width': X_d[:,3],
             'Label': Y})

table = pd.DataFrame(Iris_data_d)
table
```

```python
#Kept getting error that index 4 is out of bounds for axis 1 with size 4 with the data as it is
#So I searched online and found that all the arrays needed to be remade and reshaped


# Assigning value of one to values above the mean and zero to those below

X_new_0 = np.array( X[:,0] > X[:,0].mean(), dtype = 'int')
X_new_1 = np.array( X[:,1] > X[:,1].mean(), dtype = 'int')
X_new_2 = np.array( X[:,2] > X[:,2].mean(), dtype = 'int')
X_new_3 = np.array( X[:,3] > X[:,3].mean(), dtype = 'int')

#Reshaping the arrays

X_final0 = X_new_0.reshape(150,1)
X_final1 = X_new_1.reshape(150,1)
X_final2 = X_new_2.reshape(150,1)
X_final3 = X_new_3.reshape(150,1)
Y_d = Y.reshape(150,1)
```

```python
Iris_data_2 = OrderedDict({'Sepal Length ': X_final0[:,0],
             'Sepal Width': X_final1[:,0],
             'Petal Length': X_final2[:,0],
             'Petal Width': X_final3[:,0],
             'Label': Y})

table2 = pd.DataFrame(Iris_data_2)
table2
```

```python
#Defining new array to merge all of our arrays into one to more easily operate on
Z_d = np.hstack([X_final0,X_final1,X_final2,X_final3,Y_d])

#Better way to show what we are looking at with results
a = ['Sepal Length:','Sepal Width','Petal Length:','Petal Width:']
b = ['Setosa','Versicolor','Virginica']
Percentages = []
for i in range(4):
    for j in range(3):
        selection = np.logical_and(Z_d[:,i]>.9,Z_d[:,4] == j)
        Percentages.append(( a[i], b[j],'Selected Values:', len(Z_d[selection]),'Associated Percentage:', round((len(Z_d[selection])/len(Z_d[:,i]))*100,2)))
```

```python
# Percentages
# This better presents the data as it would normally just list out the data in a really ugly way if I called the array
for tuple in Percentages:
    print(tuple[0], tuple[1])
    print(tuple[2], tuple[3])
    print(tuple[4], tuple[5], "\n")
```
